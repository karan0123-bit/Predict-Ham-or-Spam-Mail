```python
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
```

```python
#this data is modified data
df=pd.read_csv('sms1.csv')
df.head()
```

```python
df1=df.drop('length',axis=1)
df1.head()
```

```python
df1.label=df1.label.map({'ham':0,'spam':1})
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df1=df1.dropna()
```

```python
df1.shape
```

```python
x=df1.messages #input 
y=df1.label #output
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
train_x
```

```python
train_y
```

```python
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tf=TfidfVectorizer()
```

```python
# TfidfVectorizer() library used to converts text into float form & 
# CountVectorizer() Library also some works.
# but it converts text to numeric integer form
# CountVectorizer() only words count,but TfidfVectorizer() it words count & also obtain important words score.
```

```python
p=tf.fit_transform(train_x)
```

```python
p
```

```python
j=0
for i in p:
    j+=1
    print(i,j)
```

```python
tf_train=tf.fit_transform(train_x).toarray()
```

```python
tf_train.shape
```

```python
tf_train
```

```python
tf_test=tf.transform(test_x).toarray()
tf_test
```

```python
nb=MultinomialNB()
nb.fit(tf_train,train_y)
```

```python
pred=nb.predict(tf_test)
pred
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
```

```python
accuracy_score(pred,test_y)
```

```python
f1_score(pred,test_y)
```

```python
import numpy as np
print('wrong prediction:')
print((pred!=test_y).sum(),'/',((pred!=test_y).sum()+(pred==test_y).sum()))
```

```python
cm=confusion_matrix(pred,test_y)
cm
```

```python
recall_score(pred,test_y)
```

```python
precision_score(pred,test_y)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm,annot=True,cbar=True,square=True,fmt='d')
plt.ylabel('true_label')
plt.xlabel('predicted_value')
plt.show()
```

```python
print(nb.score(tf_train,train_y))
print(nb.score(tf_test,test_y))
```
